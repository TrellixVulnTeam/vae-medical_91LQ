```python
import os
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
cachedir = 'asserts/latent_space'

out = np.load(os.path.join(cachedir, "latent_vec.npy"))
emb = np.load(os.path.join(cachedir, "latent_emb.npy"))
```

```python
print(out.shape)
print(emb.shape)
```

```python
from numpy.core.shape_base import vstack
allemb = vstack((out, emb))
print(allemb.shape)
```

```python
pca_sk = PCA(n_components=2)
vis_out = pca_sk.fit_transform(allemb)
```

```python
out = vis_out[:3200000]
emb = vis_out[3200000:]
```

```python
plt.scatter(out[:, 0], out[:, 1],marker='.')
plt.scatter(emb[:, 0], emb[:, 1],marker='o')
plt.show()
```
